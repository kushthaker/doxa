```python
user = User.query.filter(User.fully_authenticated == True)[1]
```

```python
df = labelled_focus_time_df(user)
df
```

```python
start_dt_json = {
  'year': 2020,
  'month': 3,
  'day': 8,
  'hour': 16 + 4,
  'minute': 0,
  'second': 0,
  'microsecond': 0 
}

end_dt_json = {
  'year': 2020,
  'month': 3,
  'day': 8,
  'hour': 19 + 4,
  'minute': 0,
  'second': 0,
  'microsecond': 0 
}

start_datetime = dt.datetime(**start_dt_json)
end_datetime = dt.datetime(**end_dt_json)
start_datetime, end_datetime
```

```python
mt_df = df.loc[df.datetime_utc >= start_datetime].loc[df.datetime_utc < end_datetime]
```

```python
mt_df
```

```python
crude_algo_focus_time = mt_df.is_collaborative_time == False
crude_algo_focus_time
```

```python
mt_df['crude_algo_focus_time'] = crude_algo_focus_time
mt_df
```

```python
mt_df.crude_algo_focus_time.astype(int).sum()*5 - 15 # for meeting adjustment
```

```python
gf_refocus_periods = da.focused_work_calculation(mt_df, da.gloria_mark, df_size=5)
gf_refocus_periods
```

```python
gf_refocus_periods.loc[gf_refocus_periods.notnull()]
```

```python
mt_df['focused_work_period_start_utc'] = gf_refocus_periods
refocus_times = da.get_refocus_times(mt_df)
refocus_times
```

```python
mt_df['is_refocus_time'] = refocus_times
focus_times = da.get_focus_times(mt_df)
focus_times
```

```python
mt_df['is_focus_time'] = focus_times
collaborative_times = mt_df.is_focus_time.isnull() & mt_df.is_refocus_time.isnull()
collaborative_times
```

```python
mt_df['is_collaborative_time'] = collaborative_times
mt_df
```
